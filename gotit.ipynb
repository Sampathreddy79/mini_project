```python
import pandas as pd
import os

# Define the base directory where the main folders are located
base_dir = '/content/DATA/DATA'

# List to hold individual DataFrames for each subject
data_frames = []

# Iterate through each subject folder (e.g., s01, s02)
for subject_folder in os.listdir(base_dir):
    subject_path = os.path.join(base_dir, subject_folder)

    if os.path.isdir(subject_path):
        print(f"\nProcessing subject folder: {subject_folder}")

        # Iterate through each file in the subject folder
        for file in os.listdir(subject_path):
            file_path = os.path.join(subject_path, file)

            # Check if the file is a CSV
            if file.endswith('.csv'):
                print(f"  Reading file: {file_path}")

                # Read the CSV file into a DataFrame
                data = pd.read_csv(file_path)

                # Optional: Add additional columns to identify subject and file
                data['subject_id'] = subject_folder
                data['file_name'] = file  # Helps track which file this data came from

                # Append to the list of data frames
                data_frames.append(data)

# Combine all the DataFrames if there is any data
if data_frames:
    main_data = pd.concat(data_frames, ignore_index=True)
    print("Combined Data Shape:", main_data.shape)
else:
    print("No data to concatenate. Please check the folder structure and file names.")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
import pickle

# Assuming 'combined_data' is your DataFrame

# Define features and target variables
X = combined_data[['x_acc', 'y_acc', 'z_acc', 'x_gyro', 'y_gyro', 'z_gyro']]
y_activity = combined_data['label']  # Activity labels
y_gender = combined_data['gender']    # Gender labels
y_hand = combined_data['dominant_hand']  # Dominant hand labels
y_age = combined_data['age']          # Age labels

# Split the data into training and testing sets (80% train, 20% test)
X_train, X_test, y_activity_train, y_activity_test, y_gender_train, y_gender_test, y_hand_train, y_hand_test, y_age_train, y_age_test = train_test_split(
    X, y_activity, y_gender, y_hand, y_age, test_size=0.2, random_state=42
)

# Scale the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Check the shape of the datasets
print("X_train shape:", X_train.shape)
print("y_activity_train shape:", y_activity_train.shape)
print("y_gender_train shape:", y_gender_train.shape)
print("y_hand_train shape:", y_hand_train.shape)
print("y_age_train shape:", y_age_train.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import classification_report, mean_squared_error

# 1. Train the Random Forest model for activity prediction
rf_activity = RandomForestClassifier(n_estimators=100, random_state=42)
rf_activity.fit(X_train, y_activity_train)

# Evaluate the activity model
y_activity_pred = rf_activity.predict(X_test)  # Make sure to define X_test
print("Activity Prediction Report:\n", classification_report(y_activity_test, y_activity_pred))  # Make sure to define y_activity_test

# 2. Train the Random Forest model for gender prediction
rf_gender = RandomForestClassifier(n_estimators=100, random_state=42)
rf_gender.fit(X_train, y_gender_train)  # Ensure y_gender_train is defined

# Evaluate the gender model
y_gender_pred = rf_gender.predict(X_test)  # Make sure to define X_test
print("Gender Prediction Report:\n", classification_report(y_gender_test, y_gender_pred))  # Make sure to define y_gender_test

# 3. Train the Random Forest model for dominant hand prediction
rf_hand = RandomForestClassifier(n_estimators=100, random_state=42)
rf_hand.fit(X_train, y_hand_train)  # Ensure y_hand_train is defined

# Evaluate the dominant hand model
y_hand_pred = rf_hand.predict(X_test)  # Make sure to define X_test
print("Dominant Hand Prediction Report:\n", classification_report(y_hand_test, y_hand_pred))  # Make sure to define y_hand_test

# 4. Train the Random Forest model for age prediction
rf_age = RandomForestRegressor(n_estimators=100, random_state=42)
rf_age.fit(X_train, y_age_train)  # Ensure y_age_train is defined

# Evaluate the age model
y_age_pred = rf_age.predict(X_test)  # Make sure to define X_test
print("Mean Squared Error for Age Prediction:", mean_squared_error(y_age_test, y_age_pred))  # Make sure to define y_age_test
```

```python
import pandas as pd

# Load the accelerometer data
accel_data = pd.read_csv('/Accelerometer.csv', sep='\t')  # Adjust separator if needed
print("Accelerometer Data Columns:")
print(accel_data.columns)

# Load the gyroscope data
gyro_data = pd.read_csv('/Gyroscope.csv', sep='\t')  # Adjust separator if needed
print("Gyroscope Data Columns:")
print(gyro_data.columns)

# Prepare the feature array
accel_features = accel_data[['Acceleration x (m/s^2)', 'Acceleration y (m/s^2)', 'Acceleration z (m/s^2)']]
gyro_features = gyro_data[['Gyroscope x (rad/s)', 'Gyroscope y (rad/s)', 'Gyroscope z (rad/s)']]

# Rename columns to match the model's input
accel_features.columns = ['x_acc', 'y_acc', 'z_acc']
gyro_features.columns = ['x_gyro', 'y_gyro', 'z_gyro']

# Convert columns to numeric, replacing commas and coercing errors
for col in accel_features.columns:
    accel_features[col] = pd.to_numeric(accel_features[col].astype(str).str.replace(',', '.'), errors='coerce')

for col in gyro_features.columns:
    gyro_features[col] = pd.to_numeric(gyro_features[col].astype(str).str.replace(',', '.'), errors='coerce')

# Combine the features into a single DataFrame
features = pd.concat([accel_features, gyro_features], axis=1)

# Check for NaN values
if features.isnull().values.any():
    print("Warning: NaN values found in features. Please check your data.")

# Make predictions
activity_prediction = rf_activity.predict(features)
gender_prediction = rf_gender.predict(features)
hand_prediction = rf_hand.predict(features)
age_prediction = rf_age.predict(features)

# Mapping of numerical labels to activity names
activity_map = {
    0: "Standing",
    1: "Walking",
    2: "Sitting",
    3: "Running",
    4: "Other"  # You can add more activities as needed
}

# Display predictions in a readable format
print("Activity Predictions:")
for pred in activity_prediction:
    print(activity_map[pred])

print("Gender Predictions:", ['Female' if gender == 'F' else 'Male' for gender in gender_prediction])
print("Dominant Hand Predictions:", ['Left' if hand == 'L' else 'Right' for hand in hand_prediction])
print("Predicted Age:", age_prediction)
```
